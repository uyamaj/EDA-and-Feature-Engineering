## Imputation missing values
1. Mean Value Imputation 

```python
sns.displot(df["age"], kde=True)
```

```python
#replacing null values with mean value
df["Age_mean"] = df["age"].fillna(df["age"].mean())
```

```python
df[["Age_mean", "age"]]
```

```python
#works well when we have normally distributed data
```

### 2. Median value imputation - used if there's outliers in the dataset

```python
df["age_median"]=df["age"].fillna(df["age"].median())
```

```python
df[["Age_mean", "age_median","age"]]
```

### 3. Mode Imputation Technique -used for categorical values

```python
df[df["embarked"].isnull()]
```

```python
df["embarked"].unique()
```

```python
df["embarked"].notna()
```

```python
mode_value=df[df['embarked'].notna()]['embarked'].mode()[0]
```

```python
df['embarked_mode']=df['embarked'].fillna(mode_value)
```

```python
df[['embarked_mode','embarked']]
```

```python
df["embarked_mode"].isnull().sum()
```
