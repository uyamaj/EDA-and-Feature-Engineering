```python
df_country=pd.read_csv("Country-Code.csv")
df_country.head()
```

```python
final_df=pd.merge(df, df_country,on="Country Code", how="left")
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#create pie chart. Top 3 countries that use zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.2f%%")
```

The pie chart indicates that India has the hightest number of records or transactions followed by United States and United Kingdom

```python
final_df.columns
```

```python
final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size()
```

```python
#convert to a dataframe
final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index()
```

```python
#where the column is 0 rename to Rating Count
ratings=final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

#### Observation
1. When rating is between 4.5 - 4.9 rating is excellent
2. When rating is between 4.0 - 4.4 rating is very good
3. When rating is between 3.5 - 3.9 rating is good
4. When rating is between 2.5 - 3.4 rating is average
5. When rating is between 1.8 - 2.4 rating is poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams["figure.figsize"] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count", hue="Rating color",data=ratings)
```

```python
#mapping the colors
sns.barplot(x="Aggregate rating",y="Rating Count", hue="Rating color",data=ratings,
            palette=["blue","red","orange","yellow","green","green"])
```

### Observation
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,
            palette=["blue","red","orange","yellow","green","green"])
```
